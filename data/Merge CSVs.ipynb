```python
# import required module
import os
import pandas as pd 
```

```python
# assign directory
files = []
 
for filename in os.listdir('scraped'):
    f = os.path.join('scraped', filename)
    # checking if it is a file
    if os.path.isfile(f):
        print(f)
        df = pd.read_csv(f)
        files.append(df)

for filename in os.listdir('processed/twitter'):
    f = os.path.join('processed/twitter', filename)
    if os.path.isfile(f) and filename != '.DS_Store':
        print(f)
        df = pd.read_csv(f)
        files.append(df)
        
        
        
# concatenate them all together
result = pd.concat(files, ignore_index=True)
```

```python
results_cleaned = result[["text", "scrape_time", "tweet_id", "created_at", "reply_count", "quote_count",
               "like_count", "retweet_count", "geo_full_name", "geo_id", "username", "num_followers",
               "search_keywords", "search_neighbourhood"]]
results_cleaned.head()
```

```python
results_cleaned.shape
```

```python
results_cleaned_dedupe = results_cleaned.drop_duplicates(subset='tweet_id', keep="last")
results_cleaned_dedupe.shape
```

```python
results_cleaned_dedupe.head()
```

```python
# cleaning logic 

### exclude list
bad_words = ["Trump", "Cuomo", "SF", "NYC"]

for bad in bad_words:
    df = results_cleaned_dedupe[~results_cleaned_dedupe["text"].str.lower().str.contains(bad.lower())]
```

```python
df_final = df[df['scrape_time'] < '2022-08-21']
df_final.shape
```

```python
df_final['created_at'].min()
```

```python
df_final = df_final[df_final['tweet_id'] != 1534617861154435072]
df_final = df_final[df_final['tweet_id'] != 1544376875362893824]
```
