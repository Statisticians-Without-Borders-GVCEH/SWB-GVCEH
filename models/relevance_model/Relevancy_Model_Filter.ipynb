```python
import os 
import pandas as pd 
from setfit import SetFitModel
```

```python
# iterate over files in that directory
files = pd.DataFrame(columns=['filename', 'path', 'create_time'])
directory = '/Users/sheilaflood/SWB-GVCEH/data/processed/twitter/github_actions'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    file_type = f[len(f) - 4:]
    if file_type == ".csv":
        my_list = [filename, f, os.path.getmtime(f)]
        files.loc[len(files)] = my_list
```

```python
files_sorted = files.sort_values('create_time', ignore_index=True)
files_sorted
```

```python
def relevance_model(df):
    # create the model
    print(f'Original DF Length": {len(df)}')
    model = SetFitModel.from_pretrained("sheilaflood/gvceh-setfit-rel-model2")
    all_text = df.text
    all_results = model(list[all_text])
    all_results = all_results.numpy()
    df = df[all_results == 1]  # filters out non-relevant tweets
    print(f'New DF Length": {len(df)}')
    return df
```

```python
model = SetFitModel.from_pretrained("sheilaflood/gvceh-setfit-rel-model2")
```

```python
df = pd.read_csv('/Users/sheilaflood/SWB-GVCEH/data/processed/twitter/github_actions_backup/GVCEH-tweets-combined_2023-02-17.csv')
print(len(df))
all_text = df.text
all_results = model(all_text)
all_results = all_results.numpy()
df = df[all_results == 1]  # filters out non-relevant tweets
```
