## Extracting Appendix E - Words associated with homelessness

```python
wordsdf = df.copy()

#rename column
wordsdf.rename(columns = {wordsdf.columns[1]:'word'}, inplace = True)
wordsdf
```

```python
#Finding index of the last sector name in pdf 
print(wordsdf[wordsdf['word'] == 'Poverty '].index.values)
#Finding index of the first sector name in pdf
print(wordsdf[wordsdf['word'] == 'Social problem '].index.values)
```

```python
#indexing into the df
wordsdf = wordsdf.iloc[242:277]
wordsdf
```

```python
wordsdf.drop(wordsdf.columns[0], axis = 1, inplace = True)
wordsdf
```

```python
#removing special characters like #
wordsdf['word'] = wordsdf['word'].str.replace('#', '')
wordsdf
```

```python
## separating text to multiple rows

wordsdf = wordsdf.drop('word', axis=1).join(wordsdf['word'].str.split('/', expand=True).stack().reset_index(level=1, drop=True).rename('word'))
```

```python
#reset index
wordsdf.reset_index(drop=True, inplace = True)
```

```python
wordsdf
```

```python
#saving dataframe as csv 
os.makedirs('appendixe', exist_ok=True)
wordsdf.to_csv('appendixe/words.csv') 
```
